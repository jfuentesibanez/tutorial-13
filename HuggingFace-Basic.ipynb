````python
from __future__ import print_function
import ipywidgets as widgets
from transformers import pipeline
import json
````

````python
nlp_sentence_classif = pipeline('sentiment-analysis')
````

````python
# Mock request object for local API testing
args = {
    "param": ['Such a nice weather outside !']
}
REQUEST = json.dumps({ 'args': args })
````

````python
# POST /sentence_classification
request = json.loads(REQUEST)
args = request.get('args', {})
term = args.get('param', args.get('text', None))
nlp_sentence_classif(term)
````

````python
nlp_qa = pipeline('question-answering')
````

````python
# Mock request object for local API testing
headers = {
'content-type': 'application/json'
}
body = {
'context':'Hugging Face is a French company based in New-York.' ,
'question': 'Where is based Hugging Face ?'
}
REQUEST = json.dumps({ 'headers': headers, 'body': body })
````

````python
# POST /question_answering
body = json.loads(REQUEST)['body']
# predict the output for a new sample. Function to be exposed through Titan
context = body['context']
question = body['question']

nlp_qa(context=context, question=question)
````

````python
nlp_fill = pipeline('fill-mask')
````

````python
# Mock request object for local API testing
args = {
    "param": ['Hugging Face is a French company based in ']
}
REQUEST = json.dumps({ 'args': args })
````

````python
# POST /text_generation
request = json.loads(REQUEST)
args = request.get('args', {})
term = args.get('param', args.get('text', None))

nlp_fill(term[0] + nlp_fill.tokenizer.mask_token)
````
